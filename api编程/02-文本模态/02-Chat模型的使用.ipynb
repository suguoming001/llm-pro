```python
prompt = "什么是大模型"
client = OpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url=os.getenv("DEEPSEEK_BASE_URL"),
)
max_tokens_limit = [50, 200, 1000]
for max_token in max_tokens_limit:
    response = client.chat.completions.create(
        model="deepseek-chat",
        messages=[
            {"role":"user", "content": prompt}  
        ],
        max_tokens=max_token
    )

    content = response.choices[0].message.content
    actual_token = response.usage.completion_tokens  # 实际生成的token数量
    finish_reason = response.choices[0].finish_reason  # 生成结束的原因，可能是因为达到max_tokens限制，也可能是模型认为已经完成了回答

    print(f"max_tokens={max_token}时的回答：")
    print(content)
    print(f"实际生成的token数量：{actual_token}")
    print(f"生成结束的原因：{finish_reason}")
    print("-"*50)
```

```python
questions = [
    {
        "question": "标志着中国近代史开端的历史事件是？",
        "options": ["A. 鸦片战争", "B. 洋务运动", "C. 甲午战争", "D. 辛亥革命"],
        "answer": "A"
    },
    {
        "question": "中国历史上第一个皇帝是谁？",
        "options": ["A. 夏启", "B. 嬴政", "C. 刘邦", "D. 李世民"],
        "answer": "B"
    },
    {
        "question": "鄙人不善奔跑是出自谁的口？",
        "options": ["A. 李云龙", "B. 丁伟", "C. 常乃超", "D. 孔捷"],
        "answer": "C"
    }
]
```

```python
results = []
for question in questions:

    full_question = question["question"] + "\n" + "\n".join(question["options"])
    right_answer = question["answer"]

    
    results.append({
        "full_question": full_question,
        "correct_answer": right_answer
    })

print(results)
```

```python
for i, result in enumerate(results):
    print(f"问题{i+1}：")
    print(result["full_question"])
    correct_answer = result["correct_answer"]
   
    response = client.chat.completions.create(
        model="deepseek-chat",
        messages=[
            {"role":"system", "content": "请直接给出正确选项的字母，不要任何其他的文字说明。"},
            {"role":"user", "content": result["full_question"]}  
        ],
        max_tokens=1
    )
    print(f"大模型回答：{response.choices[0].message.content}， 正确答案：{correct_answer}")
```

```python
prompt = "给我5个完全不同风格的咖啡店名字"
client = OpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url=os.getenv("DEEPSEEK_BASE_URL"),
)
temperatures = [0, 0.7, 1.9]
for temp in temperatures:
    response = client.chat.completions.create(
        model="deepseek-chat",
        messages=[
            {"role":"user", "content": prompt}  
        ],
        temperature=temp,
        max_tokens=200
    )

    content = response.choices[0].message.content

    print(f"temperature={temp}时的回答：")
    from IPython.display import display, Markdown
    display(Markdown(content))
    print("-"*50)
```
