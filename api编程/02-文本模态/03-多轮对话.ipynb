```python
from dotenv import load_dotenv
import os
from openai import OpenAI

load_dotenv(override=True)

client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY"),
    base_url=os.getenv("OPENAI_BASE_URL"),
)
```

```python
conversations = [
    {"role": "system", "content": "你是我的人工智能助手，协助我解答问题。"},
]
```

```python
# 第一轮对话
user_message_1 = "我是张三，今年64岁，明年就可以退休了，现在从事大模型算法工程师"
conversations.append({"role": "user", "content": user_message_1})

response_1 = client.chat.completions.create(
    model="gpt-5-nano",
    messages=conversations
)
```

```python
assistant_message_1 = response_1.choices[0].message.content
print(f"第一轮对话的回复：{assistant_message_1}")
conversations.append({"role": "assistant", "content": assistant_message_1})
```

```python
# 第二轮对话
user_message_2 = "你还记得我的名字、年龄和职业吗？"
conversations.append({"role": "user", "content": user_message_2})
response_2 = client.chat.completions.create(
    model="gpt-5-nano",
    messages=conversations
)
assistant_message_2 = response_2.choices[0].message.content
print(f"第二轮对话的回复：{assistant_message_2}")
conversations.append({"role": "assistant", "content": assistant_message_2})
```

```python
for conversation in conversations:
    print(f"{conversation['role']}: {conversation['content']}")
```
