```python
from dotenv import load_dotenv
import os
from openai import OpenAI


load_dotenv(override=True)

client = OpenAI(
    api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url=os.getenv("OPENROUTER_BASE_URL"),
)
```

```python
# 算术推理（Arithmetic Reasoning）
question = "Aia has 32 skittles. She eats 4 skittles and then gives half of the remaining skittles to her friend. How many skittles does she have left?"
prompt=f"""
{question}
"""
response = client.chat.completions.create(
    model="openai/gpt-3.5-turbo",
    messages=[{"role": "user", "content": prompt}]
)
print(response.choices[0].message.content)
```

```python
# 算术推理（Arithmetic Reasoning）
question = "Aia has 32 skittles. She eats 4 skittles and then gives half of the remaining skittles to her friend. How many skittles does she have left?"
prompt=f"""
Q:{question} \n
A: Let's think step by step.
"""
response = client.chat.completions.create(
    model="openai/gpt-3.5-turbo",
    messages=[{"role": "user", "content": prompt}],
    temperature=0
)
print(response.choices[0].message.content)
```

```python
# 算术推理（Arithmetic Reasoning）
question = "Aia has 32 skittles. She eats 4 skittles and then gives half of the remaining skittles to her friend. How many skittles does she have left?"
prompt=f"""
Q:{question} \n
A: Let's think step by step.
"""
response = client.chat.completions.create(
    model="openai/gpt-3.5-turbo",
    messages=[{"role": "user", "content": prompt}],
    temperature=0
)
print(response.choices[0].message.content)
print('---'*5,'COT结果','---'*5)
reasoning_path = response.choices[0].message.content

cot_prompt=f"""
Q:{question} \n
A: \n
Let's think step by step.
{reasoning_path} \n\n
Therefore, the answer is
"""
cot_response=client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": cot_prompt}],
    max_tokens=5, # 只需要提取最终数字
    temperature=0
)
print(cot_response.choices[0].message.content)
```
