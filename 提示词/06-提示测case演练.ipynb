````python
from openai import OpenAI
from dotenv import load_dotenv
import os
load_dotenv(override=True)

client = OpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url=os.getenv("DEEPSEEK_BASE_URL")
)
````

````python
def get_response(prompt):
    response = client.chat.completions.create(
        model="deepseek-chat",  
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0
    )

    return response.choices[0].message.content
````

````python
# 情感推断
infos = f"""
我需要一个新款自拍杆三脚架，具有额外的补光灯功能。\
第二天就收到货了，在快递过程中补光灯被压坏了，客服马上安排重新寄了个新的。\
但是又发现少了蓝牙遥控器，于是赶紧咨询了客服，很快又给我重新快递了一个蓝牙遥控器！\
这款自拍杆使用起来很方便，PK牌自拍杆是一家非常关心用户体验和产品极具性价比的优秀公司！
"""

prompt = f"""
以下用三个反引号分隔的产品评论的情感是什么？

评论文本: ```{infos}```
"""

print(get_response(prompt))
````

````python
# 信息提取及按格式输出
prompt = f"""
从评论文本中识别以下项目：
- 评论者购买的物品
- 制造该物品的公司

评论文本用三个反引号分隔。将你的响应格式化为以 “物品” 和 “品牌” 为键的 JSON 对象。
如果信息不存在，请使用 “未知” 作为值。
让你的回应尽可能简短。

评论文本: ```{infos}```
"""
print(get_response(prompt))
````

````python
# 信息提取
prompt = "曼联的滕嗨，就是就江湖骗子，祸害曼联，C罗、德赫亚、桑乔都被坑惨了。请提取出上述信息中的俱乐部名字(club)，主教练名字(coach)，被坑球员的中文名字(name_zh)，被坑球员的英文名字(name_eng)，并以JSON格式输出，key用英文，value用原文中的内容"
print(get_response(prompt))
````
